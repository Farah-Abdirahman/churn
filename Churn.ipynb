```python
import pandas
data = pandas.read_csv('https://modcom.co.ke/data/datasets/Churn.csv')
data.head(5)
```

```python
data.isnull().sum()
data.head(1)
```

```python
#Start encoding due to no empties
data.groupby('Contract').size()
```

```python
data['gender'].replace({'Female':0,'Male':1},inplace=True)
data['Dependents'].replace({'No':0,'Yes':1},inplace=True)
data['Contract'].replace({'One year':1,'Two year':2},inplace=True)
data['PaperlessBilling'].replace({'No':0,'Yes':1},inplace=True)
data['PhoneService'].replace({'No':0,'Yes':1},inplace=True)
data['MultipleLines'].replace({'No':0,'No Phone':1,'Yes':2},inplace=True)
data['InternetService'].replace({'DSL':0,'Fiber':1,'No':2},inplace=True)
data['gender'].replace({'Female':0,'Male':1},inplace=True)
data['OnlineSecurity'].replace({'No':0,'No Internet':1,'Yes':2},inplace=True)
data['OnlineBackup'].replace({'No':0,'No Internet':1,'Yes':2},inplace=True)
data['DeviceProtection'].replace({'No':0,'No Internet':1,'Yes':2},inplace=True)
data['TechSupport'].replace({'No':0,'No Internet':1,'Yes':2},inplace=True)
data['StreamingTV'].replace({'No':0,'No Internet':1,'Yes':2},inplace=True)
data['StreamingMovies'].replace({'No':0,'No Internet':1,'Yes':2},inplace=True)
data['Contract'].replace({'Month':0,'One Year':1,'Two Year':2},inplace=True)
data['PaymentMethod'].replace({'Bank':0,'Credit card':1,'Electronic':2,'Mailed':3},inplace=True)




```

```python
def cut(param):
  if '-' in param:
    answer = param.split('-')
    return answer[0]
  else:
    return param

cut('7590-VHVEG')
```

```python
data['customerID'] = data['customerID'].apply(cut)
data.head(1)
```

```python
newList=data[['gender','tenure','MultipleLines','OnlineBackup','DeviceProtection','TechSupport','Contract','PaperlessBilling','PaymentMethod','MonthlyCharges','Churn']]


array = newList.values
array
X = array[:,0:10]
Y = array[:,10]




# array = data.values
# array
# X = array[:,1:19]
# Y = array[:,19]

```

```python
##Synthentic minority oversampling technique-SMOTE- Solves imbalance problem.(increases minority)
# Balancing the target variable
# It aims to balance class distribution by randomly increasing minority class examples by replicating them.
# SMOTE synthesises new minority instances between existing minority instances. It generates the virtual training records by linear interpolation for the minority class.
# These synthetic training records are generated by randomly selecting one or more of the k-nearest neighbors for each example in the minority class.
# After the oversampling process, the data is reconstructed and several classification models can be applied for the processed data
#Synthetic data is artificial data that mimics real world observations and is used to train ML models when actual data is difficult or expensive to get
from imblearn.over_sampling import SMOTE
balancing = SMOTE()
newX, newY = balancing.fit_resample(X,Y)

len(newX)



#Feature Selection- dealing with datasets with vast numbers of features
```

```python
#RFE-Recursive Feature elimination
#Improves accuracy by reduction of sample datasets
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier,AdaBoostClassifier
estimator = AdaBoostClassifier(random_state=42, n_estimators= 100)
#Apply above estimator below
#here we specify 10 features, step the number of features to remove per iteration
selector = RFE(estimator, n_features_to_select=10,step = 5)
selector = selector.fit(newX, newY)
filter = selector.support_
ranking = selector.ranking_
print('Ten Best Features', filter)
print('Ranking', ranking)

newList=[['gender','tenure','MultipleLines','OnlineBackup','DeviceProtection','TechSupport','Contract','PaperlessBilling','PaymentMethod','MonthlyCharges','Churn']]

data.head(1)
```

```python
from sklearn import model_selection
X_train,X_test,Y_train,Y_test = model_selection.train_test_split(newX,newY, test_size = 0.3,random_state=42)
# Adding x_valid and Y_valid
```

```python
# #Cross Validation
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.svm import SVC #Support Vector Machine


#Cross Validation
#It is a resampling technique that uses different portions to test and train a model using diferent
# I terations
#Type1: KFOLD- Validation Technique. Used to estimate the skills of the model with the current  data we have
models = []
models.append(('DT', DecisionTreeClassifier()))
models.append(('LR', LogisticRegression(solver='liblinear', multi_class="ovr")))#ovr-one vs rest
models.append(('RF', RandomForestClassifier()))
models.append(('GB', GradientBoostingClassifier()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('GAU', GaussianNB()))
models.append(('DIS', LinearDiscriminantAnalysis()))
models.append(('SVM', SVC(gamma='auto')))


for name, model in models:
  kfold = model_selection.KFold(n_splits=10, random_state=42, shuffle=True)
  cv_results = model_selection.cross_val_score(model, X_train,Y_train,cv = kfold, scoring='accuracy')
  print(name, cv_results.mean())
```

```python
from sklearn.ensemble import RandomForestClassifier
mymodel = RandomForestClassifier()
mymodel.fit(X_train,Y_train)
```

```python
mypredictions = mymodel.predict(X_test)
print('Predicted Answers:',mypredictions)
print('Expected Answers:',Y_test)
```

```python
from sklearn.metrics import accuracy_score
print('The accuracy is',accuracy_score(mypredictions,Y_test))
```

```python
from sklearn.metrics import classification_report
print(classification_report(mypredictions,Y_test))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(mypredictions,Y_test))
#1387 no's got right while 256 were wrong
#318 yes got right while 152 got wrong
```
